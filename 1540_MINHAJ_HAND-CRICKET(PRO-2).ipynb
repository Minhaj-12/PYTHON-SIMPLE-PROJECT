```python
import random

def player_batting(player_name):
    score = 0
    while True:
        try:
            player_choice = int(input(f"{player_name}, enter your run (1-6): "))
            if player_choice not in range(1, 7):
                print("Invalid input. Enter a number between 1 and 6.")
                continue
            computer_choice = random.randint(1, 6)
            print(f"Computer chose: {computer_choice}")
            
            if player_choice == computer_choice:
                print("You're out!")
                break
            else:
                score += player_choice
                print(f"Your current score: {score}")
        except ValueError:
            print("Invalid input. Please enter a number.")
    return score

def computer_batting(target):
    score = 0
    while score < target:
        computer_choice = random.randint(1, 6)
        player_choice = int(input("Enter your run to bowl (1-6): "))
        if player_choice not in range(1, 7):
            print("Invalid input. Enter a number between 1 and 6.")
            continue
        
        print(f"Computer chose: {computer_choice}")
        if computer_choice == player_choice:
            print("Computer is out!")
            break
        else:
            score += computer_choice
            print(f"Computer's current score: {score}")
            
    return score

def main():
    print("Welcome to Hand Cricket!")
    player_name = input("Enter your name: ")
    
    # Toss
    toss_choice = input(f"{player_name}, choose Heads or Tails: ").lower()
    toss_result = random.choice(["heads","H","tails","T"])
    print(f"The toss result is: {toss_result}")
    
    if toss_choice == toss_result:
        print("You won the toss!")
        choice = input("Choose Batting or Bowling: ").lower()
    else:
        print("Computer won the toss!")
        choice = random.choice(["batting", "bowling"])
        print(f"Computer chose: {choice.capitalize()}")
    
    # Game
    if choice == "batting":
        print(f"{player_name} will bat first.")
        player_score = player_batting(player_name)
        print(f"Your final score: {player_score}")
        
        print("Computer will bat now. Try to defend your score!")
        computer_score = computer_batting(player_score)
        print(f"Computer's final score: {computer_score}")
        
    else:
        print("Computer will bat first.")
        computer_score = computer_batting(float('inf'))
        print(f"Computer's final score: {computer_score}")
        
        print(f"{player_name} will bat now. Try to chase the score!")
        player_score = player_batting(player_name)
        print(f"Your final score: {player_score}")
    
    # Result
    if player_score > computer_score:
        print(f"Congratulations {player_name}, you win!")
    elif player_score < computer_score:
        print("Computer wins! Better luck next time!")
    else:
        print("It's a draw!")
        
if __name__ == "__main__":
    main()
```
